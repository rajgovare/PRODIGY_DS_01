## **Task 1**

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Retail_sales.csv")
df.head()
```

```python
df.info()
```

```python
# Age statistics
age_stats = df['Age'].describe()
print(f"Age Statistics:\n{age_stats}")
```

```python
# Product category distribution
product_counts = df['Product Category'].value_counts()
print(f"\nProduct Category Distribution:\n{product_counts}")
```

```python
# Bar Chart for Product Category Sales
plt.figure(figsize=(10, 6))
category_sales = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Total Sales by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Total Sales (in USD)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('product_category_sales.png')
plt.show()
```

```python
# Bar Chart for Categorical Variable (Product Category)
plt.figure(figsize=(8, 5))
product_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('product_category_distribution.png')
plt.show()
```

```python
# Histogram for Continuous Variable (Age)
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('age_distribution.png')
plt.show()
```

```python
# Step 6: Gender-based Total Sales Breakdown
plt.figure(figsize=(6, 6))
gender_sales = df.groupby('Gender')['Total Amount'].sum()
sns.barplot(x=gender_sales.index, y=gender_sales.values, palette='pastel')
plt.title('Total Sales by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Total Sales (in USD)', fontsize=14)
plt.tight_layout()
plt.savefig('gender_sales_distribution.png')
plt.show()
```

```python
print("Key Insights:")
print(f"1. Highest Sales in Product Category: {category_sales.idxmax()}")
print(f"2. Average Age of Customers: {df['Age'].mean():.2f} years")
print(f"3. Gender with Highest Sales: {gender_sales.idxmax()}")
```
